```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ruta='NuevoTPIA.csv'
demanda=pd.read_csv(ruta)
demanda.info()
```

```python
demanda["Fecha"] = pd.to_datetime(demanda["Fecha"],format="%Y-%m-%d %H:%M:%S%z")
demanda.set_index("Fecha", inplace=True)
demanda
```

```python
x=demanda.index
y=demanda['SIN Imputed']
y1=demanda['T02M']
y2=demanda['RH2M']
y3=demanda['TPP6']
plt.scatter(x,y,label='DEMANDA',s=0.5)
plt.title('Curva de Demanda')
plt.xlabel('Fecha')
plt.ylabel('Demanda')
plt.legend()
plt.show()
plt.scatter(x,y1,label='T02M',alpha=0.6,s=0.5)
plt.title('T02M vs Fecha')
plt.xlabel('Fecha')
plt.ylabel('T02M')
plt.legend()
plt.show()
plt.scatter(x,y2,label='RH2M',alpha=0.6,s=0.5)
plt.title('RH2M vs Fecha')
plt.xlabel('Fecha')
plt.ylabel('RH2M')
plt.legend()
plt.show()
plt.scatter(x,y3,label='TPP6',alpha=0.6,s=0.5)
plt.title('TPP6 vs Fecha')
plt.xlabel('Fecha')
plt.ylabel('TPP6')
plt.legend()
plt.show()
x2=demanda['SIN Imputed']
plt.scatter(x2,y3,label='TPP6',alpha=0.6,s=0.5)
plt.title('Demanda vs PrecipitaciÃ³n')
plt.xlabel('Demanda')
plt.ylabel('PrecipitaciÃ³n')
plt.legend()
plt.show()
```
